```python
import cv2
import mediapipe as mp
```

```python
like_image = cv2.imread(r'D:\Courses language programming\7_Computer Vision\Computer Vision - Full Course\4-MediaPipe'
                        r'\Sign-Language-Recognition-using-mediapipe-main\like.png')
like_image = cv2.resize(like_image, (200, 180))

dislike_image = cv2.imread(r'D:\Courses language programming\7_Computer Vision\Computer Vision - Full Course\4-MediaPipe'
                           r'\Sign-Language-Recognition-using-mediapipe-main\dislike.png')
dislike_image = cv2.resize(dislike_image, (200, 180))
```

```python
finger_tips = [8, 12, 16, 20]
thump_tip = 4

mp_hand = mp.solutions.hands
hands = mp_hand.Hands()
draw_utils = mp.solutions.drawing_utils
```

```python
video = cv2.VideoCapture(0)

while True:
    _, image = video.read()
    image = cv2.flip(image, 1)
    h, w, c = image.shape

    results = hands.process(image)

    if results.multi_hand_landmarks:
        for hand_landmark in results.multi_hand_landmarks:
            lm_list = []
            for id, lm in enumerate(hand_landmark.landmark):
                lm_list.append(lm)

            finger_fold_status = []
            for tip in finger_tips:
                x, y = int(lm_list[tip].x * w), int(lm_list[tip].y * h)
                cv2.circle(image, (x, y), 15, (255, 0, 0), cv2.FILLED)

                if lm_list[tip].x < lm_list[tip-3].x:
                    cv2.circle(image, (x, y), 15, (0, 255, 0), cv2.FILLED)
                    finger_fold_status.append(True)
                else: finger_fold_status.append(False)


            if all(finger_fold_status):
                if lm_list[thump_tip].y < lm_list[thump_tip-1].y < lm_list[thump_tip-2].y:
                    print('Like')
                    cv2.putText(image, 'LIKE', (20, 30), cv2.FONT_ITALIC, 1,
                                (0, 255, 0), 3)
                    h, w, c = like_image.shape
                    image[35:35+h, 30:30+w] = like_image


                if lm_list[thump_tip].y > lm_list[thump_tip-1].y > lm_list[thump_tip-2].y:
                    print('DisLike')
                    cv2.putText(image, 'Dis-LIKE', (20, 30), cv2.FONT_ITALIC, 1,
                                (0, 255, 0), 3)
                    h, w, c = like_image.shape
                    image[35:35+h, 30:30+w] = dislike_image

            draw_utils.draw_landmarks(image, hand_landmark, mp_hand.HAND_CONNECTIONS,
                                      draw_utils.DrawingSpec((0, 0, 255), 6, 4),
                                      draw_utils.DrawingSpec((0, 255, 0), 4, 2))


    cv2.imshow('image', image)

    if cv2.waitKey(1) == ord('o'):
        break


cv2.destroyAllWindows()
video.release()
```
